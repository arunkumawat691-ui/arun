```python
import pandas as pd 
```

```python
data = pd.read_csv(r"C:\Users\Admin\Desktop\Data\sales_data_sample.csv",encoding="latin1")
df=pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
a=['ADDRESSLINE2','STATE','TERRITORY']
df[a]=df[a].fillna("unkown")
df
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'],errors='coerce')
df['ORDERDATE'] = df['ORDERDATE'].dt.strftime('%d-%m-%y')
df.head()
```

```python
df['COUNTRY']=df['COUNTRY'].str.lower().str.strip()
```

```python
df['NAME']=df['CONTACTFIRSTNAME']+df['CONTACTLASTNAME']
df
```

```python
df.columns=df.columns.str.lower().str.replace(' ','_')
df.head()
```
